# gettings tags after removing stop wrds and apply stemming for text making similar type

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
stopwords = set(stopwords.words('english'))
special_char = list("`~!@#$%^&*()-_=+*:,;.{}[]\|/'")
print(stopwords)
print(special_char)
```

```python
ps = PorterStemmer()
```

```python
def filteredText(text_str):
    text_str = str(text_str)
    str_tokesns = nltk.word_tokenize(text_str)
    rem_stop_word = []
    stem_text = []
    for w in str_tokesns:
        if w not in stopwords:
            if w not in special_char:
                rem_stop_word.append(w)
    for stm in rem_stop_word:
        stm_lower = stm.lower()
        stem_text.append(ps.stem(stm_lower))
    return ' '.join(stem_text)
```

```python
str1 = filteredText("enron methanol ; meter # : 988291\r\nthis is a follow up to the note i gave you on monday , 4 / 3 / 00 { preliminary\r\nflow data provided by daren } .\r\nplease override pop ' s daily volume { presently zero } to reflect daily\r\nactivity you can obtain from gas control .\r\nthis change is needed asap for economics purposes .")
print(str1)
```

```python
df['transformed_text'] = df['text'].apply(filteredText)
```

# setting x and y for ml model

# train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=51)
print("x_train shape => ", x_train.shape)
print("y_train shape => ", y_train.shape)

print("x_test shape => ", x_test.shape)
print("y_test shape => ", y_test.shape)
```

# KNN

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier()
```

```python
knc.fit(x_train,y_train)
knc_score = knc.score(x_test,y_test)
knc_pred = knc.predict(x_test)
print(knc_score)
print(knc_pred)
print("Accuracy scrore :- ", accuracy_score(y_test,knc_pred))
print("Confusion matrix :- \n", confusion_matrix(y_test,knc_pred))
print("precision score :- ", precision_score(y_test,knc_pred))
```
